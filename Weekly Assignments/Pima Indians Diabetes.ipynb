##2_Data Pre-Processing

```python
X=df.drop(['Outcome'], axis=1)
y=df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_rs, y_rs = smote.fit_resample(X_scaled, y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_rs, y_rs, test_size=0.25, random_state=42, stratify=y_rs
)
```

```python
print('X_train:', X_train.shape)
print('X_test:', X_test.shape)
print('y_train:', y_train.shape)
print('y_test:', y_test.shape)
```

```python
outcome_oversampling = np.concatenate([y_train, y_test])
outcome_oversampling = pd.DataFrame(outcome_oversampling)
outcome_ovs = outcome_oversampling.value_counts()
outcome_ovs .plot(kind='bar')
plt.xlabel('Outcome_OverSampling')
plt.ylabel('Count')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from keras.models import Sequential
from keras.layers import Dense
from scikeras.wrappers import KerasClassifier
from keras.optimizers import Adam
from keras.callbacks import ModelCheckpoint
```

```python
def create_model():
  model = Sequential()
  model.add(Dense(24, activation='relu', input_shape=(8,)))
  model.add(Dense(20, activation='leaky_relu'))  # Leaky ReLU
  model.add(Dense(16, activation='relu'))
  model.add(Dense(12, activation='leaky_relu'))  # Leaky ReLU
  model.add(Dense(8, activation='relu'))
  model.add(Dense(4, activation='leaky_relu'))  # Leaky ReLU
  model.add(Dense(1, activation='sigmoid'))
  model.summary()

  model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

  return model
```

```python
model = KerasClassifier(build_fn = create_model, verbose = 0)
```
